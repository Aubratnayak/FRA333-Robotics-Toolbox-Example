# Forward Kinematics for a 3R Manipulator
In this notebook, we'll compute and visualize the forward kinematics of a 3-link (3R) robot arm using the Robotics Toolbox for Python.


```python
import roboticstoolbox as rtb
from math import pi
```

## Defining the Robot
We'll begin by defining the 3R robot. Each revolute joint (R) will have a standard Denavit-Hartenberg (DH) parameter.


```python
# Define the robot using DH parameters
robot = rtb.DHRobot([
    rtb.RevoluteMDH(d=1),  # Link length = 1
    rtb.RevoluteMDH(a=1, alpha=pi/2),  # Link length = 1
    rtb.RevoluteMDH(a=1)   # Link length = 1
], name="3R_robot")
```

## Forward Kinematics
To compute the forward kinematics, we'll provide joint angles and use the robot's `fkine()` method.


```python
# Define joint angles
q = [pi/4, pi/4, pi/4]  # e.g., all joints at 45 degrees

# Compute forward kinematics
T = robot.fkine(q)

print(T)
```

## Visualization
Now, we'll visualize the robot's configuration for the given joint angles.


```python
# Plot the robot
fig = robot.plot(q, block=False)
```
